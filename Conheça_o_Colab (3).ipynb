```
#numeros total de incritos homens e mulheres (2023 e 2024)
import matplotlib.pyplot as plt
import numpy as np

# Dados
anos = ["2023", "2024"]
homens = [1.522, 1.703]
mulheres = [2.411, 2.623]

# Posições no eixo X
x = np.arange(len(anos))  # [0, 1]

# Largura das barras
largura = 0.35

# Cria o gráfico
plt.figure(figsize=(8,5))

# Barra dos homens
plt.bar(x - largura/2, homens, width=largura, color="blue", label="Homens")

# Barra das mulheres
plt.bar(x + largura/2, mulheres, width=largura, color="pink", label="Mulheres")

# Rótulos e título
plt.xlabel("Ano")
plt.ylabel("Número de Inscritos")
plt.title("Inscrições do Enem - Homens e Mulheres (2023 e 2024)")
plt.xticks(x, anos)
plt.legend()

# Mostra o gráfico
plt.show()
```

```
#inscriçoes de homens e mulheres com seu recorde (2014)
import matplotlib.pyplot as plt
import numpy as np

# Dados
anos = ["2014", "2023", "2024"]
homens_inscritos = [3.287, 1.522, 1.703]
mulheres_inscritas = [5.434, 2.411, 2.623]

# Posições no eixo X
x = np.arange(len(anos))

# Largura das barras
largura = 0.35

# Criar o gráfico
plt.figure(figsize=(9,6))

# Barras dos homens
plt.bar(x - largura/2, homens_inscritos, width=largura, color="blue", label="Homens Inscritos")

# Barras das mulheres
plt.bar(x + largura/2, mulheres_inscritas, width=largura, color="pink", label="Mulheres Inscritas")

# Rótulos e título
plt.xlabel("Ano")
plt.ylabel("Número de Inscritos")
plt.title("Inscritos do Enem por Gênero comparando com seu recorde (2014)")
plt.xticks(x, anos)
plt.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```
#numero de presentes homes e mulheres (2023 e 2024)
import matplotlib.pyplot as plt
import numpy as np

# Dados
anos = ["2023", "2024"]
homens_presentes = [1.059, 1.251]
mulheres_presentes = [1.675, 1.928]

# Posições no eixo X
x = np.arange(len(anos))  # [0, 1]

# Largura das barras
largura = 0.35

# Cria o gráfico
plt.figure(figsize=(8,5))

# Barra dos homens (azul)
plt.bar(x - largura/2, homens_presentes, width=largura, color="blue", label="Homens Presentes")

# Barra das mulheres (rosa)
plt.bar(x + largura/2, mulheres_presentes, width=largura, color="pink", label="Mulheres Presentes")

# Rótulos e título
plt.xlabel("Ano")
plt.ylabel("Número de Participantes")
plt.title("Participantes Presentes no Enem - 2023 e 2024")
plt.xticks(x, anos)
plt.legend()

# Mostra o gráfico
plt.show()
```
